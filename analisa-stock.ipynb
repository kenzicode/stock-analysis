```python
# hubungan PGAS & IDX
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import datetime
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2015,1,1)
```

```python
df = wb.DataReader(["^JKSE", "PGAS.JK", "BMRI.JK", "BBRI.JK", "UNVR.JK"], 'yahoo', start)['Adj Close']
```

```python
data = df.reset_index()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=['^JKSE', 'PGAS.JK'])
```

```python
data.isnull().sum()
```

```python
data.tail()
```

```python
plt.figure(figsize=(16,10))
plt.plot(data['^JKSE'], '-r', markersize=3)
plt.plot(data['PGAS.JK'], '-g', markersize=3)
plt.plot(data['BMRI.JK'], '-y', markersize=3)
plt.plot(data['UNVR.JK'], '-b', markersize=3)
plt.legend(['IDX','PGAS', 'BMRI', 'UNVR'], loc='upper left')
```
